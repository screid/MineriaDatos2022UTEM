```python
def filtered_data(df):

  # 1. Elimino datos faltantes para el genero
  df = df.dropna(subset=['genre'])

  # 2. Reemplazo los valores de coma milesimal
  df["votes"] = df.apply(
      lambda x:
        str(x["votes"]).replace(",",""),
        axis = 1
  )

  # 3. Transformo la columna values a numerica
  df["votes"] = df["votes"].astype('float')

  return df
```
